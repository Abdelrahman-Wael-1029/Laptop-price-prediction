```python
df.Ram = df.Ram.apply(lambda x: int(x[:-2]))
df.Weight = df.Weight.apply(lambda x: float(x[:-2]))
df['price'] = df['Price_euros']
df.drop('Price_euros', axis = 1)
df.ScreenResolution=df.ScreenResolution.apply(lambda x: x.split(' ')[-1].lower())
```

##  Preprocessing CPU


```python
intel_core_i7 = df[df['Cpu'].str.startswith('Intel Core i7')]['Cpu']
intel_core_i5 =df[df['Cpu'].str.startswith('Intel Core i5')]['Cpu']
intel_core_i3 = df[df['Cpu'].str.startswith('Intel Core i3')]['Cpu']
AMD = df[df['Cpu'].str.startswith('AMD')]['Cpu']
df = df.replace(intel_core_i7.values, 'intel core i7')
df = df.replace(intel_core_i5.values, 'intel core i5')
df = df.replace(intel_core_i3.values, 'intel core i3')
df = df.replace(df[df['Cpu'].str.startswith('Intel')]['Cpu'].values, 'intel other')
df = df.replace(AMD.values, 'AMD')
df
```

## Processing GPU

```python
Intel = df[df['Gpu'].str.startswith('Intel')]['Gpu']
AMD = df[df['Gpu'].str.startswith('AMD')]['Gpu']
ARM = df[df['Gpu'].str.startswith('ARM')]['Gpu']
Nvidia_GeForce =  df[df['Gpu'].str.startswith('Nvidia GeForce')]['Gpu']
Nvidia_Quadro =  df[df['Gpu'].str.startswith('Nvidia Quadro')]['Gpu']
Nvidia_GTX =  df[df['Gpu'].str.startswith('Nvidia GTX')]['Gpu']


df = df.replace(Intel.values, 'Intel')
df = df.replace(AMD.values, 'AMD')
df = df.replace(ARM.values, 'ARM')
df = df.replace(Nvidia_GeForce.values, 'Nvidia GeForce')
df = df.replace(Nvidia_Quadro.values, 'Nvidia Quadro')
df = df.replace(Nvidia_GTX.values, 'Nvidia GTX')

```

```python
X = df.drop('price', axis = 1)
y = df.price
```

```python
df_str = df.select_dtypes('object').columns.tolist()
df_str
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in df_str:
    X[i] = le.fit_transform(X[i])
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X, y, test_size=.23, random_state=44)
X_train.shape,X_test.shape
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

## LinearRegression

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
lg.fit(X_train,y_train)
lg_y_pred = lg.predict(X_test)
r2_score(y_test, lg_y_pred), mean_squared_error(y_test, lg_y_pred), mean_absolute_error(y_test, lg_y_pred)
```

```python
sns.scatterplot(x=y_test, y=y_test, c='red')
sns.lineplot(x=y_test, y=lg_y_pred)
```

## RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
modelRF = RandomForestRegressor()
modelRF.fit(X_train,y_train)
y_predDt = modelRF.predict(X_test)
r2_score(y_test, y_predDt), mean_squared_error(y_test, y_predDt), mean_absolute_error(y_test,y_predDt)
```

```python
sns.scatterplot(x=y_test, y=y_test, c='red')
sns.lineplot(x=y_test, y=y_predDt)
```

## DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
modelDT = DecisionTreeRegressor()
modelDT.fit(X_train,y_train)
y_predDt = modelDT.predict(X_test)
r2_score(y_test, y_predDt), mean_squared_error(y_test, y_predDt), mean_absolute_error(y_test,y_predDt)
```

```python
sns.scatterplot(x=y_test, y=y_test, c='red')
sns.lineplot(x=y_test, y=y_predDt)
```
