```python
X = df.drop('price', axis = 1)
y = df.price
```

```python
df_str = df.select_dtypes('object').columns.tolist()
df_str
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in df_str:
    X[i] = le.fit_transform(X[i])
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X, y, test_size=.23, random_state=44)
X_train.shape,X_test.shape
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
lg.fit(X_train,y_train)
lg_y_pred = lg.predict(X_test)
r2_score(y_test, lg_y_pred), mean_squared_error(y_test, lg_y_pred), mean_absolute_error(y_test, lg_y_pred)
```

```python
sns.scatterplot(x=y_test, y=y_test, c='red')
sns.lineplot(x=y_test, y=lg_y_pred)
```
